# Revisiting the Breakthrough Research papers in Generative AI in Computer Vision

# **Unsupervised representation learning using Deep Convolutional Generative Adversarial Networks (DC-GAN) by Alec Radford & co-authors**  

--> DCGANs are a type of generative model that uses a specific architecture to produce high-quality synthetic data samples that resemble a given dataset.

***********************************************
**Introduction**

This research paper discuss about the approach of addressing the challenge of learning reusable feature representations from large **unlabeled** datasets in the context of computer vision. The idea of leveraging Generative Adversarial Networks (GANs) has been introduced.

Due to mundane amount of unlabeled dataset in the Vision area which includes images and videos, here GANs can be employed to build effective image representations that can later be reused for these supervised tasks.

Authors of this paper suggested verious modification to create stable and converging GAN architecture using CNNs.
*************************************************
**The Setup:**
Here, we will be using same Generator-Discriminator concept of GANs:

 **Generator** (G): This network takes random noise (z) as input and tries to generate data (x) that resembles real data samples from a certain distribution (pdata). Essentially, it creates new data that should look like it's from the same source as the real data.

 **Discriminator** (D): The discriminator takes either real data samples (x from pdata) or generated data samples (G(z)) as input and tries to distinguish between them. It essentially acts as a binary classifier, determining whether the input is from the real data distribution or the generated distribution.


 **Architecture guidelines for stable Deep Convolutional GANs**

  * Replace any pooling layers with strided convolutions (discriminator) and fractional-strided convolutions (generator).

  * Use batchnorm in both the generator and the discriminator.

  * Remove fully connected hidden layers for deeper architectures.

  * Use ReLU activation in generator for all layers except for the output, which uses Tanh.

  * Use LeakyReLU activation in the discriminator for all layers.

Later, In the paper, Authors have explained about the details of learned model:

* Image Classification with Trained Discriminators: They demonstrate the utility of the trained discriminator networks for image classification tasks. They show that the discriminators can be used as feature extractors, achieving competitive performance compared to other unsupervised learning algorithms.

* Visualization of Filters: They present a visualization of the filters learned by GANs. They empirically show that specific filters have learned to represent specific objects or patterns in the data. This provides insight into the internal representations learned by the network.

* Vector Arithmetic Properties of Generators: They highlight the intriguing vector arithmetic properties exhibited by the generators. These properties allow for easy manipulation of various semantic qualities of the generated samples. This means that by performing arithmetic operations on the generator's input noise vector, one can control and modify certain aspects of the generated images. Its same as image to image  translation applications.
